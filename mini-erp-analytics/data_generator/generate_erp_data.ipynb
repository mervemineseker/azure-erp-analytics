```python
# ===============================
# MINI ERP DATA GENERATOR
# ===============================

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import os

# klas√∂r
BASE_DIR = "erp_output"
os.makedirs(BASE_DIR, exist_ok=True)

np.random.seed(42)

# -------------------------------
# MASTER DATA
# -------------------------------
customers = pd.DataFrame({
    "customer_id": range(1, 101),
    "customer_name": [f"Customer {i}" for i in range(1, 101)],
    "country": np.random.choice(["ES", "DE", "FR", "IT"], 100)
})

products = pd.DataFrame({
    "product_id": range(1, 51),
    "product_name": [f"Product {i}" for i in range(1, 51)],
    "unit_cost": np.round(np.random.uniform(5, 50, 50), 2),
    "unit_price": np.round(np.random.uniform(20, 120, 50), 2)
})

# -------------------------------
# SALES ORDERS
# -------------------------------
orders = []
start_date = datetime(2025, 1, 1)

for i in range(1, 5001):
    order_date = start_date + timedelta(days=random.randint(0, 364))
    product = products.sample(1).iloc[0]
    qty = random.randint(1, 20)

    orders.append({
        "order_id": i,
        "order_date": order_date.date(),
        "customer_id": random.randint(1, 100),
        "product_id": product.product_id,
        "quantity": qty,
        "unit_price": product.unit_price,
        "order_value": round(qty * product.unit_price, 2)
    })

sales_orders = pd.DataFrame(orders)

# -------------------------------
# INVOICES
# -------------------------------
invoices = sales_orders.copy()
invoices["invoice_id"] = invoices["order_id"]
invoices["invoice_date"] = invoices["order_date"] + pd.to_timedelta(
    np.random.randint(1, 10, len(invoices)), unit="D"
)

# -------------------------------
# PAYMENTS
# -------------------------------
payments = invoices.sample(frac=0.85)
payments = payments[["invoice_id", "invoice_date", "order_value"]]
payments["payment_date"] = payments["invoice_date"] + pd.to_timedelta(
    np.random.randint(5, 40, len(payments)), unit="D"
)
payments.rename(columns={"order_value": "amount"}, inplace=True)

# -------------------------------
# SAVE FILES
# -------------------------------
customers.to_csv(f"{BASE_DIR}/customers.csv", index=False)
products.to_csv(f"{BASE_DIR}/products.csv", index=False)
sales_orders.to_csv(f"{BASE_DIR}/sales_orders.csv", index=False)
invoices.to_csv(f"{BASE_DIR}/invoices.csv", index=False)
payments.to_csv(f"{BASE_DIR}/payments.csv", index=False)

print("‚úÖ ERP DATA CREATED")
print("üìÅ Folder:", BASE_DIR)
```
